```python
def aapClusters(sequences, indexing, mmm=1, delimeter='*'):
    l = len(sequences[0])
    clusters = []
    masks = itertools.combinations(np.arange(l), mmm)
    for mask in masks:
        mask = [-1] + list(mask) + [l]
        masker = lambda x: delimeter.join([x[mask[i] + 1:mask[i + 1]] for i in range(mmm + 1)])
        factor = set([masker(x) for x in sequences])
        c = {m:[] for m in factor}
        for i, x in enumerate(sequences):
            c[masker(x)].append(indexing[i])
        for m in factor:
            if len(c[m]) > 1:
                clusters.append(c[m])
    return clusters

def EdgeListfromClusters(n, clusters):
    edges = set([])
    for cluster in clusters:
        for x, y in itertools.combinations(cluster, 2):
            edges.add((x, y))
    return list(edges)
```

```python
def GetClusters(cdr3):
    edges = []
    cdr3_length = np.array([len(x) for x in cdr3])
    for l in range(6, 20):
        indexes = np.arange(cdr3.shape[0])[cdr3_length == l]
        if indexes.shape[0] > 0:
            a = cdr3[indexes]
            clusters = aapClusters(a, indexes)
            e = EdgeListfromClusters(cdr3.shape[0], clusters)
            edges += e
    
    graph = nx.Graph()
    graph.add_edges_from(edges)
    components = list(nx.connected_components(graph))
    return components
```

```python
def ClusterIndexes(array):
    factor = np.array(list(set(array)))
    element2i = {x:i for i, x in enumerate(factor)}
    indexes = [[] for x in factor]
    for i, x in enumerate(array):
        indexes[element2i[x]].append(i)
    return indexes
```

```python
epitops = ClusterIndexes(data_pd['epitope'])
for indexes in epitops:
    if len(indexes) > 1:
        acdr3 = data_pd['alpha.cdr3'].values[indexes]
        bcdr3 = data_pd['beta.cdr3'].values[indexes]

        alpha_clusters = GetClusters(acdr3)
        beta_clusters = GetClusters(bcdr3)

        a_component = -1 + np.zeros((acdr3.shape[0]), dtype=int)
        for i, x in enumerate(alpha_clusters):
            a_component[list(x)] = i
        b_component = -1 + np.zeros((bcdr3.shape[0]), dtype=int)
        for i, x in enumerate(beta_clusters):
            b_component[list(x)] = i
        try:
            t = np.zeros((len(alpha_clusters), len(beta_clusters)), dtype=int)
            for i in range(acdr3.shape[0]):
                t[a_component[i], b_component[i]] += 1
            if np.max(t) > 1 and t.size > 1:
                x = np.sum(t, axis=0)
                y = np.sum(t, axis=1)
                e = np.dot(y.reshape(-1, 1), x.reshape(1,-1)) / np.sum(t)

                plt.scatter(t, e)
                plt.plot([0, np.max(t)], [0, np.max(t)])
            plt.show()
        except:
            1
```

```python
aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', '_']
aa2num_table = {aa[i]:i for i in range(21)}

def aa2num(aa):
    return aa2num_table[aa]

def protein2array(protein):
    return np.array([aa2num(aa) for aa in protein])

def array2protein(array):
    return ''.join([aa[x] for x in array])
```

```python
def standartize_cdr3_length(protein):
    insertion_places = {6:3, 7:4, 8:4, 9:5, 10:5, 11:6, 12:6, 13:7, 14:7, 15:8}
    insertion_place = insertion_places[len(protein)]
    insertion_length = 15 - len(protein)
    return protein[:insertion_place] + '_' * insertion_length + protein[insertion_place:]
```

```python
def MutualInformationTable(A, B):
    t = np.zeros((A.shape[1], B.shape[1]))
    for i in range(A.shape[1]):
        for j in range(B.shape[1]):
            t[i, j] = adjusted_mutual_info_score(A[:, i], B[:, j])
    return t
```

```python
acdr3 = data_pd['alpha.cdr3'].values
bcdr3 = data_pd['beta.cdr3'].values
A = np.zeros((data_pd.shape[0], 15), dtype=object)
B = np.zeros((data_pd.shape[0], 15), dtype=object)
for i in range(data_pd.shape[0]):
    for j in range(15):
        A[i, j] = acdr3[i][j]
        B[i, j] = bcdr3[i][j]
```

```python
def GetPairSampling(A, B):
    assert A.shape[0] % 2 == 0
    X = np.zeros((2 * A.shape[0], A.shape[1] + B.shape[1]))
    X[::2, :A.shape[1]] = A
    X[1::2, :A.shape[1]] = A
    X[::2, A.shape[1]:] = B
    X[1::4, A.shape[1]:] = B[1::2]
    X[3::4, A.shape[1]:] = B[::2]
    
    y = np.zeros((2 * A.shape[0]))
    y[::2] = 1
    
    return X, y
```
